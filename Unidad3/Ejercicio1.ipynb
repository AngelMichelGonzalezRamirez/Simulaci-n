##**Unidad3 - Ejercicio1:**
####**Contexto:** Un distribuidor de artículos electrónicos recibe varios pedidos diarios. Cada día, el distribuidor tiene una probabilidad de éxito de cumplir un pedido (es decir, de tener en stock los artículos solicitados) del 70%. Desea simular la cantidad de pedidos diarios que logra satisfacer para gestionar adecuadamente sus inventarios y minimizar las faltantes de stock.

####**Objetivo:** Generar una variable aleatoria que simule el número de pedidos satisfechos por día durante una semana, considerando que cada día el distribuidor recibe 20 pedidos.

**Maestro: José Gabriel Rodríguez Rivas**

**Alumno: Ángel Michel González Ramírez**


```python
import numpy as np
import matplotlib.pyplot as plt

def simular_pedidos_satisfechos(dias=7, pedidos_por_dia=20, probabilidad_exito=0.7):
    # Generamos datos usando la distribución binomial:
    pedidos_satisfechos = np.random.binomial(n=pedidos_por_dia, p=probabilidad_exito, size=dias)

    # Mostramos resultados:
    print("Pedidos satisfechos por día:", pedidos_satisfechos)

    # Grafica de los resultados:
    plt.bar(range(1, dias + 1), pedidos_satisfechos, color='skyblue')
    plt.xlabel('Dias')
    plt.ylabel('Pedidos Satisfechos')
    plt.title('Simulación de Pedidos Satisfechos por Día')
    plt.xticks(range(1, dias + 1))
    plt.show()

# Ejecutar simulación
simular_pedidos_satisfechos()
```
